```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))
        except ValueError:
            print("Please enter a valid number.")
            continue

        if amount < 50:
            print("Invalid: Minimum recharge is ₹50")
        elif amount not in valid_plans:
            print("Invalid: Plan not available")
        else:
            print("Recharge Successful!")
            break

validate_recharge()
```

```python
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: Reorder Alert")
        else:
            print(f"{item}: Stock OK")


# Example usage
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 20
}

check_inventory(inventory)
```

```python
def student_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example usage
marks_list = [60, 45, 70, 55]
result = student_result(marks_list)
print("Result:", result)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
def check_eligibility(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_eligibility(attendance_list)
print("Attendance Status:", status)
```

```python
def check_password(password):
    if len(password) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Example usage
pwd = input("Enter password: ")
print(check_password(pwd))
```
